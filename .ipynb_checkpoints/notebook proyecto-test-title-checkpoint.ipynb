```python
import pandas as pd
import numpy as np
import re
import nltk
import pickle
import os
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import HashingVectorizer
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
dff = pd.read_csv('Datasets/fake-news/Fake.csv', usecols = ['title'])
dft = pd.read_csv('Datasets/fake-news/True.csv', usecols = ['title'])
dff = dff.dropna()
dft = dft.dropna()
dff['class'] = 0
dft['class'] = 1
```

```python
frames = [dff,dft]
dfn = pd.concat(frames)
dfn.info
```

```python
text_list = dfn['title'].tolist()
count = CountVectorizer()
bag = count.fit_transform(text_list)
```

```python
print(bag.toarray())
```

```python
tfidf = TfidfTransformer(use_idf=True,norm='l2',smooth_idf=True)
np.set_printoptions(precision=2)
print(tfidf.fit_transform(count.fit_transform(text_list)).toarray())
```

```python
def preprocessor(text):
    text = (re.sub('[\W]+', ' ', text.lower()))
    return text
```

```python
dfn['title'] = dfn['title'].apply(preprocessor)
```

```python
dfn.head()
```

```python
def tokenizer(text):
    return text.split()
```

```python
porter = PorterStemmer()
lemmatizer = WordNetLemmatizer() 
def tokenizer_porter(text):
    return [lemmatizer.lemmatize(word) for word in text.split()]
```

```python
stop = stopwords.words('english')
```

```python
X_train, X_test, y_train, y_test = train_test_split(dfn['title'], dfn['class'], test_size=0.50, random_state=1)
```

```python
X_train = X_train.to_numpy()
y_train = y_train.to_numpy()
X_test = X_test.to_numpy()
y_test = y_test.to_numpy()
```

```python
tfidf = TfidfVectorizer(strip_accents=None,lowercase=False,preprocessor=None)
```

```python
param_grid = [
{'vect__ngram_range': [(1,1)],
'vect__stop_words': [stop, None],
'vect__tokenizer': [tokenizer,tokenizer_porter],
'clf__penalty': ['l1', 'l2'],
'clf__C': [1.0, 10.0, 100.0]},
{'vect__ngram_range': [(1,1)],
'vect__stop_words': [stop, None],
'vect__tokenizer': [tokenizer,tokenizer_porter],
'vect__use_idf':[False],
'vect__norm':[None],
'clf__penalty': ['l1', 'l2'],
'clf__C': [1.0, 10.0, 100.0]}]
```

```python
lr_tfidf = Pipeline([('vect', tfidf),('clf',LogisticRegression(random_state=0,solver='liblinear'))])
```

```python
gs_lr_tfidf = GridSearchCV(lr_tfidf, param_grid,
scoring='accuracy',
cv=5, verbose=2,
n_jobs=1)
```

```python
gs_lr_tfidf.fit(X_train, y_train)
```

```python
print('Best parameter set: %s ' % gs_lr_tfidf.best_params_)
```

```python
print('CV Accuracy: %.3f' % gs_lr_tfidf.best_score_)
clf = gs_lr_tfidf.best_estimator_
print('Test Accuracy: %.3f'% clf.score(X_test, y_test))
```
